```python
# supprime les colonnes de  'x_m', 'y_m' et set lindex par defaut a la colonne de nombre de forages ('nfor') 
data=rawdata.copy()
data=data.drop(['x_m', 'y_m'], axis=1)
data.head()
```

```python
# convertion de la colonne debit en categories (0: forage sec, 0-1 m3/h, 1-3 m3/h, 3-6 m3/h , 6-10 m3/h, +10 m3/h)
data['debitm'].map(lambda x:x+1)
def cat_debitm(debitm):
    if debitm==0:
        return ' forage sec'
    elif debitm>0 and debitm<=3: 
        return '0.-3 m3/h'
    elif debitm>3 and debitm<=6: 
        return '3.-6 m3/h'
    elif debitm >6 and debitm<=10: 
        return '6.-10 m3/h'
    else : 
        return '+10. m3/h'
data['debitm']=data['debitm'].map(cat_debitm)
data.set_index('nfor')
data.tail(15)
```

```python
#ANALYSE EN COMPOSANTE PRINCIPALE 
df=data.copy() 
df['forme'] = df['forme'].astype('category').cat.codes
df['type'] = df['type'].astype('category').cat.codes
df['geol'] = df['geol'].astype('category').cat.codes
df=df.drop('naEau', axis=1)
df.set_index('nfor', inplace=True)
df.head()
```

```python
#defenir les composantes X et y 
# split data table into data X and class labels y

X = df.iloc[:,0:7].values
y = df.iloc[:,7].values
```

```python
from sklearn.preprocessing import StandardScaler
X_std = StandardScaler().fit_transform(X)
```

```python
import numpy as np
mean_vec = np.mean(X_std, axis=0)
cov_mat = (X_std - mean_vec).T.dot((X_std - mean_vec)) / (X_std.shape[0]-1)
print('Covariance matrix \n%s' %cov_mat)
```

```python
print('NumPy covariance matrix: \n%s' %np.cov(X_std.T))
```

```python
cov_mat = np.cov(X_std.T)

eig_vals, eig_vecs = np.linalg.eig(cov_mat)

print('Eigenvectors \n%s' %eig_vecs)
print('\nEigenvalues \n%s' %eig_vals)
```

```python
cor_mat1 = np.corrcoef(X_std.T)

eig_vals, eig_vecs = np.linalg.eig(cor_mat1)

print('Eigenvectors \n%s' %eig_vecs)
print('\nEigenvalues \n%s' %eig_vals)
```

```python
cor_mat2 = np.corrcoef(X.T)

eig_vals, eig_vecs = np.linalg.eig(cor_mat2)

print('Eigenvectors \n%s' %eig_vecs)
print('\nEigenvalues \n%s' %eig_vals)
```

```python
u,s,v = np.linalg.svd(X_std.T)
u
```

```python
for ev in eig_vecs:
    np.testing.assert_array_almost_equal(1.0, np.linalg.norm(ev))
print('Everything ok!')

# Make a list of (eigenvalue, eigenvector) tuples
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]

# Sort the (eigenvalue, eigenvector) tuples from high to low
eig_pairs.sort()
eig_pairs.reverse()

# Visually confirm that the list is correctly sorted by decreasing eigenvalues
print('Eigenvalues in descending order:')
for i in eig_pairs:
    print(i[0])
    
```

```python
matrix_w = np.hstack((eig_pairs[0][1].reshape(7,1), 
                      eig_pairs[1][1].reshape(7,1)))

print('Matrix W:\n', matrix_w)
```
