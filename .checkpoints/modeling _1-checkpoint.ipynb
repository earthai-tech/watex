```python
# convertion de la colonne debit en categories (0: forage sec, 0-1 m3/h, 1-3 m3/h, 3-6 m3/h , 6-10 m3/h, +10 m3/h)
df=df_raw.copy()
df['debitm'].map(lambda x:x+1)
def cat_debitm(debitm):
    if debitm==0:
        return 'forage sec'
    elif debitm>0 and debitm<=3: 
        return '0.-3 m3/h'
    elif debitm>3 and debitm<=6: 
        return '3.-6 m3/h'
    elif debitm >6 and debitm<=10: 
        return '6.-10 m3/h'
    else : 
        return '+10. m3/h'
df['debitm']=df['debitm'].map(cat_debitm)
df.set_index('nfor', inplace=True)
df.head()
```

# modeling 

```python
#constructions des pipelines 

numerical_features = make_column_selector(dtype_include=np.number)
categorical_features= make_column_selector(dtype_exclude=np.number)
# numerical_features = ['puiss', 'ampl','IFS', 'surfOhm']
# categorical_features=[ 'forme', 'type','geol']#SelectKBest(f_classif, k=3)

numerical_pipeline=make_pipeline(PolynomialFeatures(10, include_bias=False), SelectKBest(f_classif, k=4), RobustScaler())
# numerical_pipeline=make_pipeline(RobustScaler())
categorical_pipeline = make_pipeline(OneHotEncoder())

preprocessor =make_column_transformer((numerical_pipeline,numerical_features ),
                        (categorical_pipeline,categorical_features))  
```

```python
#construction de modeles pour voir lequels donne un bon resultats 
model_dtc = make_pipeline (preprocessor, DecisionTreeClassifier(random_state=0))
model_sgd = make_pipeline (preprocessor, SGDClassifier(random_state=0))
model_knn = make_pipeline (preprocessor, KNeighborsClassifier(n_neighbors=10, metric='manhattan'))
model_svm = make_pipeline (preprocessor, SVC(C=100, gamma=1e-3, random_state=0))
model_rand= make_pipeline (preprocessor, RandomForestClassifier(n_estimators=200, random_state=0))
model_ada=make_pipeline (preprocessor, AdaBoostClassifier(random_state=0))

dict_of_models={'KNN': model_knn,
                'SVM': model_svm,
                'RandomForestClassifier': model_rand,
                'AdaboostClassifier': model_ada }

#model_5=VotingClassifier(("SGDC", model_1),
       
#                        ("KNN", model_2),
#                         ('DTC', model_3),
#                         ('SVC', model_4), voting='hard')                      
```

```python
#Essayer de visualiser tout ceci en creer une fonction evaluation 

def evaluation(model): 
    
    model.fit(X_train, y_train)
#     model.score(X_test, y_test)
    y_pred = model.predict(X_test)
    
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    N, train_score, val_score=learning_curve(model, X_train, y_train,
                                            cv=4, scoring='accuracy',
                                            train_sizes=np.linspace(0.2,1,10))
    plt.figure(figsize=(12,8))
    plt.plot(N, train_score.mean(axis=1),label='train score')
    plt.plot(N, val_score.mean(axis=1),label='val_score')
    plt.legend()
    
```

```python
for key, value in dict_of_models.items(): 
    print(key)
    evaluation(value)
```
